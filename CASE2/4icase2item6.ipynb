```
tons_exported=tons_exported[0:23]
mean=tons_exported.mean()
range=tons_exported.max()-tons_exported.min()
tons_exported_normalized=(tons_exported-tons_exported.mean())/(tons_exported.max()-tons_exported.min())
```

```
X=normalized[0:23]
Y=tons_exported_normalized
```

```
len(X)
```

```
X_train=X[0:int(0.9*len(X))]
Y_train=Y[0:int(0.9*len(Y))]
X_test=X[int(0.9*len(X)):]
Y_test=Y[int(0.9*len(Y)):]
Y_test
```

```
X_train_tensor=torch.tensor(X_train.values,dtype=torch.float)
Y_train_tensor=torch.zeros(Y_train.shape[0],1)
Y_train_tensor[:,0]=torch.tensor(Y_train.values,dtype=torch.float)
X_test_tensor=torch.tensor(X_test.values,dtype=torch.float)
Y_test_tensor=torch.zeros(Y_test.shape[0],1)
Y_test_tensor[:,0]=torch.tensor(Y_test.values,dtype=torch.float)
print(X_train_tensor.shape)
print(Y_train_tensor.shape)
print(X_test_tensor.shape)
print(Y_test_tensor.shape)
print(Y_test)
```

```
Train=TensorDataset(X_train_tensor,Y_train_tensor)
Test=TensorDataset(X_test_tensor,Y_test_tensor)
```

```
train_loader=DataLoader(Train,batch_size=X_train_tensor.shape[0],shuffle=False)
test_loader=DataLoader(Test,batch_size=X_test_tensor.shape[0],shuffle=False)
```

```
class nn_model(nn.Module):
  def __init__(self,input_size,hidden_size):
    super(nn_model,self).__init__()
    self.input_size=input_size
    self.hidden_size=hidden_size
    self.layer1=nn.Linear(self.input_size,self.hidden_size[0])
    self.layer2=nn.Linear(self.hidden_size[0],self.hidden_size[1])
    self.layer3=nn.Linear(self.hidden_size[1],self.hidden_size[2])
    self.layer4=nn.Linear(self.hidden_size[2],self.hidden_size[3])
    self.layer5=nn.Linear(self.hidden_size[3],1)
    
    self.relu1=nn.LeakyReLU(0.9)
    self.relu2=nn.LeakyReLU(0.9)
    self.relu3=nn.LeakyReLU(0.9)
    
  def forward(self,x):
    out=self.layer1(x)
    #out=self.relu1(out)
    out=self.layer2(out)
    #out=self.relu2(out)
    out=self.layer3(out)
    #out=self.relu3(out)
    out=self.layer4(out)
    
    out=self.layer5(out)
    #out=self.relu(out)
    
    return out
```

```
hidden=[64,32,16,8]
model=nn_model(X_train_tensor.shape[1],hidden)

cost=nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
#optimizer=torch.optim.SGD(model.parameters(),lr=0.001)
```

```
epochs=100
training_loss=[]
test_loss=[]
e=0
while e<epochs:

  print(e+1,"/",epochs)
  loss_epoch_train=0.0
  loss_epoch_test=0.0
  for x,y in train_loader:
    model.train()
    yhat=model(x)
    loss=cost(yhat,y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    loss_epoch_train+=loss.item()
  for x,y in test_loader:
    model.eval()
    yhat=model(x)
    loss=cost(yhat,y)
    loss_epoch_test+=loss.item()
  print("train loss:",loss_epoch_train/len(train_loader),"test loss:",loss_epoch_test)
  training_loss.append(loss_epoch_train/len(train_loader))
  test_loss.append(loss_epoch_test)
  e+=1
```

```
fig=go.Figure()
fig.add_trace(go.Scatter(y=training_loss,mode='lines',name='train'))
fig.add_trace(go.Scatter(y=test_loss,mode='lines',name='test'))

fig.show()
```
